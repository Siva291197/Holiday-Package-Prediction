```python
# Let's fetch the outlier indices of Duration of Pitch, Number of Followups, Number of Trips & Monthly Income.

package_outliers = package.copy()

def outlier_indices(df,features):
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    
    IQR = q3 - q1
    
    lower_fence = q1 - (1.5 * IQR)
    upper_fence = q3 + (1.5 * IQR)
    
    outlier_indices_list = df[(df[feature] < lower_fence) | (df[feature] > upper_fence)].index
    
    return outlier_indices_list
```

```python
index_list = []

outlier_features = ['DurationOfPitch','NumberOfFollowups','NumberOfTrips','MonthlyIncome']

for feature in outlier_features:
    index_list.extend(outlier_indices(package_outliers,feature))
```

```python
# No. of indices that has outliers (with duplicates).
len(index_list)
```

```python
# No of. indices that has outliers (without duplicates).
len(set(index_list))
```

```python
# Function to remove the indices that has outliers.

def remove_outliers(df,idx_list):
    sorted_index = sorted(set(idx_list))
    df = df.drop(sorted_index)
    return df
```

```python
# Shape of dataset before removing outliers.

package_outliers.shape
```

```python
# Call remove_outliers function to remove the outliers based on index.

package_outliers = remove_outliers(package_outliers,index_list)
```

```python
# Function to check features that are correlated to each other.

def corr_func(dataset,threshold):
    corr_features = set()
    corr_matrix = dataset.corr()
    
    for i in range(len(corr_matrix)):
        for j in range(i):
            if round(corr_matrix.iloc[i,j],2) > threshold:
                corr_features.add(corr_matrix.columns[i])
#                 corr_features.add(corr_matrix.columns[j])
    return corr_features
```

```python
# Call corr_func by setting threshold value as 0.85

corr_func(X,0.85)
```

```python
# Applying standardization for numerical features and one hot encoding for categorical features using column transformer

numerical_scaling = StandardScaler()
categorical_encoding = OneHotEncoder(drop = 'first', dtype = int, handle_unknown='ignore')

preprocess = ColumnTransformer([
    ('Encoder', categorical_encoding,categorical_features),
    ('Standardizer', numerical_scaling,numerical_features)
])
```

```python
preprocess
```

```python
X_train.head()
```

```python
# Fit the training data.

preprocess.fit(X_train)
```

```python
# Transform the training & Test data. Output will be in arrays.

X_train_preprocess = preprocess.transform(X_train)
X_test_preprocess  = preprocess.transform(X_test)
```

## Multimodel Training.

```python
models = {
    "DummyClassifier" : DummyClassifier(),
    "LogisticRegression" : LogisticRegression(),
    'DecisionTreeClassifier' : DecisionTreeClassifier(),
    "RandomForestClassifier" : RandomForestClassifier(),
    "AdaBoost"               : AdaBoostClassifier()
}

list(models.keys())
```

```python
models = {
    "DummyClassifier" : DummyClassifier(),
    "LogisticRegression" : LogisticRegression(),
    'DecisionTreeClassifier' : DecisionTreeClassifier(),
    "RandomForestClassifier" : RandomForestClassifier(),
    "AdaBoost"               : AdaBoostClassifier()
}

for i in range(len(list(models))):
    #models.values() - will return models in a dictionary.
    clf_model = list(models.values())[i]
    #Training the models.
    clf_model.fit(X_train_preprocess,y_train)
    
    #Make predictions.
    y_train_pred = clf_model.predict(X_train_preprocess)
    y_test_pred  = clf_model.predict(X_test_preprocess)
    
    #Training set performance.
    train_accuracy     = accuracy_score(y_train,y_train_pred)
    train_precision    = precision_score(y_train,y_train_pred)
    train_recall       = recall_score(y_train,y_train_pred)
    train_f1_score     = f1_score(y_train,y_train_pred,average='weighted')
    train_rocauc_score_negative_class = roc_auc_score(y_train,clf_model.predict_proba(X_train_preprocess)[:,0])
    train_rocauc_score_positive_class = roc_auc_score(y_train,clf_model.predict_proba(X_train_preprocess)[:,1])
    
    #Test Set performance.
    test_accuracy      = accuracy_score(y_test,y_test_pred)
    test_precision     = precision_score(y_test,y_test_pred)
    test_recall        = recall_score(y_test,y_test_pred)
    test_f1_score      = f1_score(y_test,y_test_pred,average='weighted')
    test_rocauc_score_negative_class  = roc_auc_score(y_test,clf_model.predict_proba(X_test_preprocess)[:,0])
    test_rocauc_score_positive_class  = roc_auc_score(y_test,clf_model.predict_proba(X_test_preprocess)[:,1])
               
    print(list(models.keys())[i])
    print('-'*len(list(models.keys())[i]))
    print('Training Performance')
    print(f'-Accuracy : {round(train_accuracy,4)}')
    print(f'-Precision : {round(train_precision,4)}')
    print(f'-Recall : {round(train_recall,4)}')
    print(f'-F1 Score : {round(train_f1_score,4)}')
    print(f'-ROC AUC Score for Negative Class : {round(train_rocauc_score_negative_class,4)}')
    print(f'-ROC AUC Score for Positive Class : {round(train_rocauc_score_positive_class,4)}')
    print('-Confusion Matrix:')
    print(confusion_matrix(y_train,y_train_pred))
    print('-Classification Report:')
    print(classification_report(y_train,y_train_pred))
    
    print('Testing Performance')
    print(f'-Accuracy : {round(test_accuracy,4)}')
    print(f'-Precision : {round(test_precision,4)}')
    print(f'-Recall : {round(test_recall,4)}')
    print(f'-F1 Score : {round(test_f1_score,4)}')
    print(f'-ROC AUC Score for Negative Class : {round(test_rocauc_score_negative_class,4)}')
    print(f'-ROC AUC Score for Positive Class : {round(test_rocauc_score_positive_class,4)}')
    print('-Confusion Matrix:')
    print(confusion_matrix(y_test,y_test_pred))
    print('-Classification Report:')
    print(classification_report(y_test,y_test_pred))
    print('='*35)
    print('\n')
```

## Insights:

Decision Tree Classifier can be adopted for this use case when compared to Logistic Regression/Random Forest/AdaBoost even though the accuracy of random Forest is high.

### Reason:
1. For this use case, the value of FN (Type II Error) should be as low as possible for positive class(1). Because, if the model falsely predicts that the customer is not willing to opt for a package wherein he/she is actually willing to opt.
2. However, FP (Type I Error) can be compromised in positive class as this won't be impacting the business.

## Hyper Parameter Tuning for Decision Tree Classifier using CV.

```python
# Applying GridSearch CV.

grid_params = {
    'criterion' : ['gini','entropy','log_loss'],
    'splitter'  : ['best','random'],
    'max_depth' : [5,10,15,20,None],
    'min_samples_split' : [2,4,6,8,10,12,14,None],
    'max_features' : ['sqrt','log2','None']
}
```

```python
grid_cv = GridSearchCV(estimator=DecisionTreeClassifier(),param_grid=grid_params,scoring = 'accuracy',n_jobs=-1,cv = StratifiedKFold(n_splits = 5))
grid_cv
```

## Points to Note:
1. n_splits = 5 - denotes that k = 5 folds. So, a training data of 3120 will be splitted as 624 datapoints for each fold.
2. Stratified is used to avoid class imbalance between training and validation set.
3. n_jobs = -1 - denotes that all the processors should be used when performing grid search cv.

```python
# Train the grid cv model.
grid_cv.fit(X_train_preprocess,y_train)
```

```python
# Evaluate best training accuracy score.

grid_cv.best_score_
```
